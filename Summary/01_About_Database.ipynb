### DB

 - Database
 - 데이터를 통합하여 관리하는 데이터의 집합
 
----
#### DBMS

 - Database Management System
 - 데이터베이스를 관리하는 미들웨어 시스템을 데이터베이스 관리시스템
 
     ex) mysql : RDBMS / mongo DB : NoSQL

---
#### RDBMS

 - Relational Database Management System
 
ex) Oracle(유료), Mysql(무료, oracle이 인수), postgresql, sqlite
 
 - 테이블 사이에 키 값으로 관계를 가지고 있는 데이터 베이스 (merge 가능)
 - 데이터 사이의 관계 설정으로 최적화된 스키마를 설계 가능
 - Mysql 학습
 
---
### NoSQL

 ex) mongoDB , Hbase, cassandra
 
 - 테이블 사이에 관계가 없이 저장하는 데이터 베이스 (merge 불가능)
 - 데이터 사이에 관계가 없으므로 복합성은 줄고, 많은 데이터 저장 가능
 - key, value의 dictionary 데이터 형태로 구성
 - MongoDB학습
 
---
#### Feature

 - 다양한 운영체제 사용가능 , 다양한 프로그래밍 언어 지원
 - 표준 SQL 사용
 - 세계적으로 많이 사용
 - 작고 강력하고 가격이 저렴

### RDBMS

#### Feature 

 - 데이터 분류, 정렬, 탐색속도가 빠름
 - 신뢰도가 높음
 - 스키마 수정이 어려움
 
 - Server > Database > Table > Row > Value
 
 [**row**, tuple, record], [column, **field**, attribute]
 
 
#### Table
 
 - row, column 으로 이루어져있는 데이터 베이스의 기본 단위
 
#### column 

 - 세로축 데이터
 
#### row

 - 가로축 데이터 
 
#### value

 - row, column에 포함되어 있는 데이터
 
#### key

 - row 식별자
 
#### relationship

 - 1:1
 - 1:N
 - N:M

#### schema

 - 데이터 베이스의 구조를 만드는 디자인

----
### NoSQL

 - Not only SQL
 
 - RDBMS의 한계를 극복하기 위해 만들어진 데이터
 - 확장성이 좋음 : 데이터의 분산처리 용이
 - 데이터 저장이 유연함 : 테이블의 연결이 없기 때문에 구조 변경 불필요
 - schema , join 없음
 - collecion 관계가 없기 떄문에 모든 데이터가 있어야함
 - key, value 형태의 JSON 포맷 사용
 - select : RDBMS > NoSQL , insert : RDBMS < NoSQL
 - transaction을 지원하지 않음 (동시 수정에 대한 신뢰성 지원 안됨)
 
 - Server > Database > Collection > Document > key:Value
 
#### mongodb

 - C++로 작성된 오픈소스 데이터 베이스
 - 뛰어난 확장성
 - NoSQL에서는 가장 인지도가 높음
 
#### collection

 - RDBMS의 Table과 같은 개념
 
#### document

 - RDBMS의 Row와 같은 개념
 
#### key

 - RDBMS의 column과 같은 개념
